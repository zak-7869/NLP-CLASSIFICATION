```python
df = df.rename(columns={'tpoic': 'topic'})
```

```python
df = df.dropna(subset=['comment'])
df['comment'] = df['comment'].astype(str)
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
ps = PorterStemmer()

corpus=[]
for i in range(len(df)):
    rp=re.sub('[^a-zA-Z]'," ",df['comment'][i])
    rp=rp.lower()
    rp=rp.split()
    rp=[ps.stem(word) for word in rp if not word in stopwords.words('english')]
    rp=" ".join(rp)
    corpus.append(rp)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer(max_features=None)
X=cv.fit_transform(corpus).toarray()
X
```

```python
X.shape
```

```python
from sklearn.preprocessing import LabelEncoder
le= LabelEncoder()
df["topic"]=le.fit_transform(df[["topic"]])
df["topic"]
```

```python
y=df["topic"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test= train_test_split(X,y,train_size=0.8,random_state=True)
```

```python
from sklearn.naive_bayes import MultinomialNB
model=MultinomialNB()
model.fit(X_train,y_train)
```

```python
ypred_train=model.predict(X_train)
from sklearn.metrics import accuracy_score
accuracy_score(y_train,ypred_train)
```

```python
ypred_test=model.predict(X_test)
from sklearn.metrics import accuracy_score
accuracy_score(y_test,ypred_test)
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(model,X,y,cv=5).mean()
```
